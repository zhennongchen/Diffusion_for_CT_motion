```python
data_path = '/mnt/camca_NAS/Portable_CT_data'
patient_sheet = pd.read_excel(os.path.join(data_path,'Patient_list', 'NEW_CT_concise_collected_portable_motion_w_fixed_CT_info_edited.xlsx'),dtype={'Patient_ID': str, 'Patient_subID': str, 'Patient_ID_fixed':str, 'Patient_subID_fixed':str})
patient_sheet['use_portable'] = patient_sheet['use_portable'].fillna(0)
patient_sheet = patient_sheet[(patient_sheet['use_portable'] != 0) & (patient_sheet['use_portable'] != 'no')]
print(patient_sheet.shape)
```

```python
Result = []

for i in range(0,patient_sheet.shape[0]):
    row = patient_sheet.iloc[i]
    patient_id = row['Patient_ID']
    patient_subid = row['Patient_subID']
    batch = 5
    motion_name = 'random_1'
    
    if batch == 5:
        train = 'test'
    else:
        train = 'train'

    print(patient_id, patient_subid)
       
    # motion-free
    motion_free_file = os.path.join(data_path,'nii_imgs_202404/motion',patient_id,patient_subid,'portable/img_resample_avg.nii.gz')
    print(motion_free_file)
    # assert os.path.isfile(motion_free_file) == 1
    motion_free_ds_file = ''

    # motion image
    motion_corrupted_file = os.path.join(data_path,'nii_imgs_202404/motion',patient_id,patient_subid,'portable/img_resample_avg.nii.gz')
    print(motion_corrupted_file)
    # assert os.path.isfile(motion_corrupted_file) == 1

    # motion image downsampled and cropped
    motion_corrupted_ds_file = ''
            
    # motion parameters
    motion_parameter_file = ''

    # PAR
    PAR_folder = ''

    # PAR file
    PAR_file = ''

    # apply true MVF to PAR
    image_from_PAR_MVF = ''      

    Result.append([train, batch, patient_id, patient_subid, motion_name, 0, 50, motion_free_file, motion_free_ds_file, motion_corrupted_file, 
                                motion_corrupted_ds_file, motion_parameter_file, PAR_file, image_from_PAR_MVF])      


column_list = ['group','batch','PatientID','AccessionNumber','MotionName','PAR_start_slice', 'PAR_end_slice', 'MotionFree_File', 'MotionFreeDs_File',
                'Motion_File', 'MotionDs_File', 'MotionParam','PAR_File','ImageFromPARMVF']
df = pd.DataFrame(Result, columns = column_list)
df.to_excel(os.path.join('/mnt/camca_NAS/diffusion_ct_motion/data','Patient_list/Patient_list_real_portable_CT_202404_resample_avg.xlsx'),index=False)
```
